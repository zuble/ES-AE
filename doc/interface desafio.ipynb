```python
from PyQt5.QtCore import *
from PyQt5.QtWidgets import *
from PyQt5.QtGui import *

from qt_material import apply_stylesheet
import time

import sys
import numpy as np
```

```python
class my_widget(QWidget):
    """
    This is my widget.
    """
    def __init__(self):
        super().__init__()
        self.setWindowTitle('Interface do desafio')
        
        
         # variables
        # count variable
        self.count = 0
  
        # start flag
        self.flag = False
        
        self.global_layout = QGridLayout(self)
        
       #label nomeo estado da comunicacao
        self.state_n_label=QLabel(self)
        self.state_n_label.setText('Estado da comunicação:')
        self.global_layout.addWidget(self.state_n_label,0,0,1,1,Qt.AlignCenter)
        
        #label do estado
        self.statelabel=QLabel(self)
        self.statelabel.setText('Ativa') #por a variar de acordo com uma variavel
        self.global_layout.addWidget(self.statelabel,1,0,1,1,Qt.AlignCenter)
        self.statelabel.setStyleSheet("border: 1px solid blue;")
        
        #label nome da localizacao
        self.loc_n=QLabel(self)
        self.loc_n.setText('Localização do veículo:')
        self.global_layout.addWidget(self.loc_n,2,0,1,1,Qt.AlignCenter)
        
        #label das coordenadas
        self.loc=QLabel(self)
        self.loc.setText('x=1 y=2') #ir buscar a variavel que vira do mapeamento
        self.global_layout.addWidget(self.loc,3,0,1,1,Qt.AlignCenter)
        self.loc.setStyleSheet("border: 1px solid blue;")
        
        #add close button
        self.closeButton=QDialogButtonBox(QDialogButtonBox.Close)
        self.global_layout.addWidget(self.closeButton,6,2,1,1,Qt.AlignRight)
        self.closeButton.clicked.connect(self.close)
        
          
        #add a child layout
        self.childlayout=QGridLayout(self)
        self.global_layout.addLayout(self.childlayout,0,3,5,2)
        
        #adicionar estado da bateria
        
        #label do estado da bateria
        self.bate_n=QLabel(self)
        self.bate_n.setText('Estado da bateria:')
        self.childlayout.addWidget(self.bate_n,0,0,1,1)
        
        #nr da bateria
        self.bate=QProgressBar(self)
        self.bate.setMaximum(100)
        self.bate.setMinimum(0)
        self.bate.setValue(50)
        self.childlayout.addWidget(self.bate,0,1,1,1)
        self.bate.setStyleSheet("QProgressBar::chunk "
                  "{"
                    "background-color: green;"
                  "}")
        
        #self.perc=QLabel(self)
        #self.perc.setText('%')
        #self.childlayout.addWidget(self.perc,0,2,1,1)
        
        #add a second child layout
        self.child2layout=QHBoxLayout(self)
        self.childlayout.addLayout(self.child2layout,1,1,1,2)
        
        #label do tempo de operaçao
        self.tempo_n=QLabel(self)
        self.tempo_n.setText('Tempo de operacao:')
        self.childlayout.addWidget(self.tempo_n,1,0,1,1)
        
        #label do tempo que passou
        self.tempo=QLabel(self)
        self.tempo.setText(str(self.count))
        self.child2layout.addWidget(self.tempo)
        
        self.s=QLabel(self)
        self.s.setText('s')
        self.child2layout.addWidget(self.s)
        
        #botoes da contagem
        self.time_button = QPushButton(self)
        self.time_button.setText('Iniciar contagem do tempo')
        self.childlayout.addWidget(self.time_button,2,0,1,2)
        
        self.stop_button=QPushButton(self)
        self.stop_button.setText('Parar Contagem')
        self.childlayout.addWidget(self.stop_button,3,0,1,2)
        
        self.reset_button=QPushButton(self)
        self.reset_button.setText('Reset')
        self.childlayout.addWidget(self.reset_button,4,0,1,2)
    
        self.timer=QTimer(self)
        self.timer.timeout.connect(self.ShowTime)
        self.timer.start(100)
        #ativacao dos botoes
        self.time_button.clicked.connect(self.Start)
        self.stop_button.clicked.connect(self.Stop)
        self.reset_button.clicked.connect(self.Reset)
        
    def Start(self):
        self.flag=True
    def Stop(self):
        self.flag=False
    def Reset(self):
        self.flag=False
        self.count=0
        self.tempo.setText(str(self.count))
    def ShowTime(self):
        if self.flag:
            self.count+=1
        text=str(self.count/10)
        self.tempo.setText(text)
        
    
    
#The next piece of code is fundamental to run it on Jupyter, not so much on other IDE's
#basically there can only be one pyqt5 application running for an interpreter
#and it does not die simply by closing the app, so we need to check if there is already an application available

qapp = QCoreApplication.instance()
if qapp is None:
    qapp = QApplication(sys.argv)
    
if __name__ == "__main__": 
    #start the widget
    ui = my_widget()
    apply_stylesheet(qapp, theme='light_blue.xml')
    #show the widget
    ui.show()
    #start the event loop
    qapp.exec_()   
```
